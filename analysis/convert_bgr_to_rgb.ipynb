```python
# 一开始的save_CAM里面的jetcolormap的通道保存错了，导致之前训练的代码全是保存成了BGR格式的图片，现在把这些图片转换成RGB格式。
# 新的代码已经改正了这个问题。
```

```python
from pathlib import Path

input_path = "/work/SKIING/chenkaixu/code/ClinicalGait-CrossAttention_ASD_PyTorch/logs/train/3dcnn_attn_map_True_none_bgr_channel/2025-11-12/17-56-55/test_all_feature_maps"
input_path = Path(input_path)

output_path = input_path.parent / "test_all_feature_maps_rgb"

```

```python
import cv2
from PIL import Image

for one_img in input_path.rglob("*.png"):
    if one_img.suffix.lower() not in [".png"]:
        continue

    # 1️⃣ 读取图片（OpenCV默认是BGR）
    # 原图保存的就是bgr，所以这里直接读就行，cv2.imread读出来的就是bgr格式，相当于转换了一下
    img_bgr = cv2.imread(str(one_img))

    # 2️⃣ 转换为RGB
    img_rgb = img_bgr  # 或者：cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)

    # 3️⃣ 保存为RGB图（用Pillow最稳妥）
    output_img_path = one_img.with_name(one_img.stem + "_rgb" + one_img.suffix)
    output_img_path = output_path / one_img.relative_to(input_path)
    output_img_path.parent.mkdir(parents=True, exist_ok=True)
    Image.fromarray(img_rgb).save(output_img_path)

```
